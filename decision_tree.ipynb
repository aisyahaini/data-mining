```python
#import library
import numpy as np
import pandas as pd
from sklearn import tree
```

```python
#import dataset
irisDataset = pd.read_csv('Dataset Iris.csv', delimiter=';', header=0)
```

```python
#Menampilkan beberapa baris pertamaa dari dataset iris
irisDataset.head()
```

```python
#Mengubah kelas (kolom 'species') dari string ke unique-integer
irisDataset["Species"] = pd.factorize(irisDataset.Species)[0]
```

```python
#Menampilkan beberapa baris pertamaa dari dataset iris setelah kolom species diubah
irisDataset.head()
```

```python
#Menampilkan dataset Iris
print(irisDataset)
```

```python
irisDataset = irisDataset.drop(labels="Id", axis=1)
```

```python
#Menampilkan dataset iris setelah menghapus kolom id
print(irisDataset)
```

```python
#Mengubah dataframe ke array numpy
#iriDataset = irisDataset.as_matrix()
irisDataset = irisDataset.to_numpy()
```

```python
#Menampilkan dataset iris setelah diubah ke array
print(irisDataset)
```

```python
#Membagi dataset, 40 baris data untuk training
#dan 20 baris data untuk testing
dataTraining = np.concatenate((irisDataset[0:40, :], irisDataset[50:90, :]),
                              axis=0)
dataTesting = np.concatenate((irisDataset[40:50, :], irisDataset[90:100, :]),
                             axis=0)
```

```python
#Menampilkan data testing
print(dataTesting)
len(dataTesting)
```

```python
#Memecah dataset ke input dan label
inputTraining = dataTraining[:, 0:4]
inputTesting = dataTesting[:, 0:4]
labelTraining = dataTraining[:, 4]
labelTesting = dataTesting[:, 4]
```

```python
#Mendefinisikan decision tree classifier
model = tree.DecisionTreeClassifier()
```

```python
#Mentraining model
model = model.fit(inputTraining, labelTraining)
```

```python
#Memprediksi input data testing
hasilPrediksi = model.predict(inputTesting)
print("Label sebenarnya ", labelTesting)
print("Hasil prediksi: ", hasilPrediksi)
```

```python
#Menghitung akurasi
prediksiBenar = (hasilPrediksi == labelTesting).sum()
prediksiSalah = (hasilPrediksi != labelTesting).sum()
print("Prediksi Benar :", prediksiBenar, "data")
print("Prediksi Salah :", prediksiSalah, "data")
print("Akurasi :", prediksiBenar/(prediksiBenar+prediksiSalah) * 100, "%")
```
